# Segment Wise Analysis of Sales, Discount and Profit

Maximum sales is in Home Office segement

Maximum profit is seen in Home Office segment

# State wise Profit Analysis

```python
data_state=Sample_Superstore.groupby(['State'])[['Profit','Discount']].mean()
data_state
```

```python
data_state.sort_values('Profit')[['Profit']].plot(kind='bar',figsize=(10,10))
plt.title('State wise Profit',fontsize=15)
plt.ylabel('Profit Amount')
plt.xlabel('States')
plt.show()
```

# State wise Discount Analysis

```python
data_state.sort_values('Discount')[['Discount']].plot(kind='bar',figsize=(10,10))
plt.title('State wise Discount',fontsize=15)
plt.ylabel('Discount Amount')
plt.xlabel('States')
plt.show()
```

```python
#Determine Loss States
data_state[data_state['Profit']<0]
```

Above are the states which are sufering from Loss along with the discount provided to such states

# City Wise Sales, Discount and Profit Analysis

```python
data_city=Sample_Superstore.groupby(['City'])[['Sales','Discount','Profit']].mean()
data_city=data_city.sort_values(['Profit'])
data_city
```

# Top 15 cities in terms of Profit

```python
data_city['Profit'].tail(15).plot(kind='bar',figsize=(10,10),color='blue')
plt.title('Top 15 Cities in terms of Profit', fontsize=15)
plt.ylabel('Total Profit')
plt.xlabel('Cities')
plt.show()
```

# 15 cities with least profits

```python
data_city['Profit'].head(15).plot(kind='bar',figsize=(10,10),color='red')
plt.title('Top 15 Cities in terms of Loss', fontsize=15)
plt.ylabel('Total Loss')
plt.xlabel('Cities')
plt.show()
```

# Sub-Category Wise Analysis of Sales, Discount and Profit

```python
plt.figure(figsize=(15,15))
plt.pie(data_subcategory['Sales'],labels=data_subcategory.index, autopct='%1.1f%%')
plt.title('Sub-Category Wise Sales',fontsize=20)
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
